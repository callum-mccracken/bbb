# Boost and Rotate, but only with 4 jets

```python
import tools
from four_jet_network import FourJetNetwork

filename = 'user.jagrundy.20736236._000001.MiniNTuple.root'
```

```python
# open
s_table = tools.open_file(filename, sort_by="tag")
```

```python
print("filtering from", len(s_table), "total events")

# filter so we get events with 4 jets, 3 or 4 b-jets, and 3 tags
j4 = (s_table.njets == 4) # 4 jets
b34 = (s_table.nbjets == 3) | (s_table.nbjets == 4) # 3 or 4 b-jets
nt3 = s_table.nbtags==3  # 3 b-tags
events = s_table[j4 & b34 & nt3]

# and ensure that the 3 tags are actually correct
# this results in very little event loss
events = events[events.truth[:,0] == 1]
events = events[events.truth[:,1] == 1]
events = events[events.truth[:,2] == 1]

n_events = len(events)
print(n_events)
```

```python
# pad events out to 4 events
events = tools.pad(events, length=4)
# boost/rotate
events = tools.boost_and_rotate(events)
```

Create FourJetNetwork

```python
nn = FourJetNetwork(events)
```

```python
nn.learn()
```

```python
nn.evaluate()
```

At this point we have a network that classifies sets of 4 jets where 3 are true b-jets and one is uncertain. It seems to do decently well! Try evaluating this model using a larger dataset.

```python
# evaluate this on the set where we don't check if tags are correct
events2 = s_table[j4 & b34 & nt3]
events2 = tools.pad(events2, length=4)
events2 = tools.boost_and_rotate(events2)
nn.evaluate(events2)
```

```python
# evaluate on the set where we don't require that there are 3-4 b-jets
events3 = s_table[j4 & nt3]
events3 = tools.pad(events3, length=4)
events3 = tools.boost_and_rotate(events3)
nn.evaluate(events3)
```

```python
# evaluate on the set with >=4 jets, not ==4
events4 = s_table[s_table.njets>=4 & nt3]
events4 = tools.pad(events4, length=4)
events4 = tools.boost_and_rotate(events4)
nn.evaluate(events4)
```
